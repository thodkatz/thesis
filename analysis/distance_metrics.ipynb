## Dataset exploration

```python
from thesis import SAVED_RESULTS_PATH
import os
import matplotlib.pyplot as plt
from thesis.datasets import NaultPipeline
import scanpy as sc
import pertpy as pt
from seaborn import clustermap
from typing import List, Optional



FIGURES_PATH = SAVED_RESULTS_PATH / "figures"
os.makedirs(FIGURES_PATH, exist_ok=True)
```

### Nault

```python
nault = NaultPipeline()
nault_data = nault.dataset
```

```python
sc.pp.pca(nault_data)
sc.pp.neighbors(nault_data)
sc.tl.umap(nault_data)
```

```python
DISTANCE_METRICS = ['edistance', 'wasserstein', 'euclidean', 'mean_pairwise', 'mmd'] 
def distance_plots(data, groupby: str, dataset_name: str, groups: Optional[List] = None, distance_metric = DISTANCE_METRICS):
    for distance_metric in DISTANCE_METRICS:
        distance = pt.tl.Distance(distance_metric, obsm_key="X_pca")
        df = distance.pairwise(data, groupby=groupby,groups=groups)
        clustermap(df, robust=True, figsize=(10, 10))
        plt.savefig(FIGURES_PATH / f'{dataset_name}_{distance_metric}_clustermap.pdf', dpi=300, bbox_inches='tight')
        plt.show()
```

```python
groups = [0, 0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
distance_plots(nault_data, groups=groups, groupby='Dose', dataset_name='nault')
```

```python
hepatocytes = nault_data[nault_data.obs['celltype'] == 'Hepatocytes - portal']
groups = [0, 0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
distance_plots(hepatocytes, groups=groups, groupby='Dose', dataset_name='hepatocytes')
```

```python
hepatocytes_central = nault_data[nault_data.obs['celltype'] == 'Hepatocytes - central']
groups = [0, 0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
distance_plots(hepatocytes_central, groups=groups, groupby='Dose', dataset_name='hepatocytes_central')
```

```python
dose_highest = nault_data[nault_data.obs['Dose'] == 30]
liver_cell_types = [
    'Hepatocytes - central',
    'Hepatocytes - portal',
    'Cholangiocytes',
    'Stellate Cells',
    'Portal Fibroblasts',
    'Endothelial Cells'
]
distance_plots(dose_highest, groups=liver_cell_types, groupby='celltype', dataset_name='dose_highest')
```

```python
dose_lowest = nault_data[nault_data.obs['Dose'] == 0.01]
liver_cell_types = [
    'Hepatocytes - central',
    'Hepatocytes - portal',
    'Cholangiocytes',
    'Stellate Cells',
    'Portal Fibroblasts',
    'Endothelial Cells'
]
distance_plots(dose_lowest, groups=liver_cell_types, groupby='celltype', dataset_name='dose_lowest')
```
